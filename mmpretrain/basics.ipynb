## 基于配置文件进行训练
- 对训练模式进行了统一

- 模型配置文件
    - 结构: 方便手改
    - `type`: 如ImageClassifier
    - `backbone` 图像特征提取, 主干网络可以相同
    - `neck` 衔接
    - `head` 最后输出
- 数据集配置
- 规划配置
    - 如优化器, 混合精度训练 等
- 结果配置
    - `default_hooks.checkpoint.save_best = True` 可以自己添加, 总是保存最好的结果


```python
from mmengine import Config

cfg = Config.fromfile("./configs/resnet/resnet18_8xb32_in1k.py")
cfg.model  # 非常方便修改, 因为直接是dict
```
